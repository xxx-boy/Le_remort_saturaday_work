### ☆Pythonで行う、比率の検定

pythonを使って母比率の統計検定量を求める場合は以下のようになります。
与えられた設問より、１が出た回数、サンプル数、標本比率、母比率をそれぞれ用意します。
次に、母比率における統計検定量zを求める式を使って検定量zの値を求めます。
次に、有意水準を使い、statsのnorm.ppfを使用して、z値を求めて、if文で検定することができます。
```python

import math

x = 1500 #1が出た回数
n = 10000 #サンプル数
P_hat = x/n #標本比率
P_0 = 1/6 #母比率

z = (P_hat - P_0)/math.sqrt(P_0*(1-P_0)/n)
print("母比率の統計検定量 = ", z)
母比率の統計検定量 = -4.472135954999579

from scipy import stats
alpha = 0.05 #有意水準

z_value = stats.norm.ppf(1-alpha/2)

if -z_value < z < z_value:
    print("z値は", z_value, "なので、帰無仮説を採択し、対立仮説を棄却するという結果になり、",
          "このサイコロの1が出る確率は1/6ではないとは言えず、このサイコロは1の出やすさに関して歪んでいるとは言えないと結論づけられます。")
    
else:
    print("z値は", z_value, "なので、帰無仮説を棄却し、対立仮説を採択するという結果になり、",
          "このサイコロの1が出る確率は1/6ではなく、このサイコロは1の出やすさに関して歪んでいると結論づけられます。")
```

````python
# 問12
# (24)
# 今回は標準的な検定の中の平均の差の検定を行う、またその場合のtは以下のように求める
````

$$
    母分散が等しい時の平均の差の検定統計量\\
    t  = \frac{\bar{x_1} + \bar{x_2}}{u\sqrt{\frac{1}{n_1}+\frac{1}{n_2}}} \\
    この時、u^2は \\
    u^2 = \frac{(n_1-1)u^2_1 + (n_2-1)u^2_2 }{(n_1-1)+(n_2-1)} \\
    よって、
    t = \frac{(6-1)×\frac{1}{6-1}(13.549) + (6-1)×\frac{1}{6-1}(7.763)}{(6 - 1)+(6 - 1)} \\
    となり、これを解くと \\
